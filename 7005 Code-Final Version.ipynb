# Univariate analysis

```python
def plot_categorical_distribution(column_name):
    plt.figure(figsize=(10, 4))
    
    plt.subplot(1, 2, 1)
    sns.countplot(y=column_name, data=df, palette='coolwarm')  
    plt.title(f'Distribution of {column_name}')
    
    ax = plt.gca()
    for p in ax.patches:
        ax.annotate(f'{p.get_width()}', (p.get_width(), p.get_y() + p.get_height() / 2), 
                    ha='center', va='center', xytext=(10, 0), textcoords='offset points')
    
    sns.despine(left=True, bottom=True)
    
    plt.subplot(1, 2, 2)  
    df[column_name].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('coolwarm'), startangle=90, explode=[0.05] * df[column_name].nunique())  
    plt.title(f'Percentage Distribution of {column_name}')  
    plt.ylabel('')    
      
    # Display the plots  
    plt.tight_layout()  
    plt.show()  

# for the 'gender' column
plot_categorical_distribution('Gender')
```

# Numerical variables (emotional variables)

```python
# Function to perform univariate analysis for numeric columns
def univariate_analysis(data, column, title):
    plt.figure(figsize=(10, 6))

    plt.subplot(1, 2, 1)
    sns.histplot(data[column], kde=True, color='teal')
    plt.title(f'{title} Distribution with KDE')

    plt.subplot(1, 2, 2)
    sns.boxplot(data[column], color='coral')
    plt.title(f'{title} Boxplot')
    plt.tight_layout()
    plt.show()

    print(f'\nSummary Statistics for {title}:\n', data[column].describe())

columns_to_analyze = ['Nervous/Anxious', 'Worrying', 'Trouble Relaxing ', 'Easily Annoyed',
       'Excessive Worry ', 'Restless', 'Fearful ', 'Anxiety Value', 'Upset',
       'Lack of Control', 'Nervous/Stress ', 'Inadequate Coping', 'Confident',
       'Things Going Well', 'Control Irritations', 'Top Performance',
       'Angered by Bad Performance', 'Overwhelmed', 'Stress Value',
       'Lack of Interest', 'Feeling Down', 'Sleep Problems', 'Tired',
       'Appetite Issues', 'Self-Doubt', 'Concentration Issues',
       'Movement Issues', 'Suicidal Thoughts', 'Depression Value']

for column in columns_to_analyze:
    univariate_analysis(df, column, column.replace('_', ' '))
```

# Model and Model testing

```python
df.info()
```

```python
#label category
from sklearn.preprocessing import LabelEncoder
categorical_columns = ['Age', 'Gender','University','Department','Academic Year',
                       'Current CGPA','Waiver/Scholarship','Anxiety Label','Stress Label','Depression Label'] 

encoder = LabelEncoder()
for i in categorical_columns:
    df[i] = encoder.fit_transform(df[i])
    

df.info()

```

```python
#select variables

x = df.drop(columns=["Anxiety Label","Anxiety Value", "Stress Label", "Stress Value",  "Depression Value", "Depression Label"])
y_anxiety = df["Anxiety Label"]
y_Stress = df["Stress Label"]
y_Depression = df["Depression Label"]

print(x.shape)  
print(y_anxiety.shape) 
print(y_Stress.shape)
print(y_Depression.shape)
```

```python
from sklearn.model_selection import train_test_split
#Training and Testing Dataset
X_train, X_test, y_train_Anxiety, y_test_Anxiety = train_test_split(x, y_anxiety, test_size=0.2, random_state=42)
X_train, X_test, y_train_Stress, y_test_Stress = train_test_split(x, y_Stress, test_size=0.2, random_state=42)
X_train, X_test, y_train_Depression, y_test_Depression = train_test_split(x, y_Depression, test_size=0.2, random_state=42)

print(X_train.shape, y_train_Depression.shape)
```

```python
#Model Training-Decision Tree
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.tree import DecisionTreeClassifier

Tree = DecisionTreeClassifier(criterion='gini', random_state=42,class_weight='balanced')

Tree.fit(X_train, y_train_Anxiety)
y_predict_anxiety = Tree.predict(X_test)

cm = confusion_matrix(y_test_Anxiety,y_predict_anxiety)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title("Confusion Matrix for Anxiety Prediction")
plt.xlabel("Predicted Labels")
plt.ylabel("Actual Labels")
plt.show()

report = classification_report(y_test_Anxiety, y_predict_anxiety, output_dict=True)
report_df = pd.DataFrame(report).transpose()
print(report_df)

feature_importances = Tree.feature_importances_
feature_names = X_train.columns 
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})

importance_df = importance_df.sort_values(by='Importance', ascending=False)
print(importance_df)

```

```python
Tree = DecisionTreeClassifier(criterion='gini', random_state=42)

Tree.fit(X_train, y_train_Stress)
y_predict_Stress = Tree.predict(X_test)

cm = confusion_matrix(y_test_Stress,y_predict_Stress)

report = classification_report(y_test_Stress, y_predict_Stress, output_dict=True)
report_df = pd.DataFrame(report).transpose()
print(report_df)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title("Confusion Matrix for Stress Prediction")
plt.xlabel("Predicted Labels")
plt.ylabel("Actual Labels")
plt.show()

feature_importances = Tree.feature_importances_

print(feature_importances)
```

```python
Tree = DecisionTreeClassifier(criterion='gini', random_state=42,class_weight='balanced')

Tree.fit(X_train, y_train_Depression)
y_predict_Depression = Tree.predict(X_test)

cm = confusion_matrix(y_test_Depression,y_predict_Depression)

report = classification_report(y_test_Depression, y_predict_Depression, output_dict=True)
report_df = pd.DataFrame(report).transpose()
print(report_df)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title("Confusion Matrix for Depression Prediction")
plt.xlabel("Predicted Labels")
plt.ylabel("Actual Labels")
plt.show()

feature_importances = Tree.feature_importances_
feature_names = X_train.columns 
importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})

importance_df = importance_df.sort_values(by='Importance', ascending=False)
print(importance_df)
```
